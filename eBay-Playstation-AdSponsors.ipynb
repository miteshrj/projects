## Loop to download the first 10 pages

```python
import requests
import time
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
import collections as cl
import json
import codecs
import warnings
import mysql.connector
```

```python
for x in range(1, 11):
    
    ## Setting the url code to be save the first 10 search pages from eBay for the search of 'Playstation 4 Slim'
    ## Use %02d to get the int in a 2 digit format
    url = "https://www.ebay.com/sch/i.html?_from=R40&_nkw=playstation+4+slim&_sacat=0&_ipg=100&rt=nc&LH_BIN=1&_pgn="+"%02d" %x
    
    print(url)
    source = requests.get(url, headers = {'user-agent':'Mozilla/5.0'})
    file_name = "ebay_ps4_"+"%02d"%x+".htm"
    save_file = open(file_name, "wb")
    save_file.write(source.content)
    save_file.close()
    
    print("Search page "+str(x)+" saved as file: "+file_name+"\n")
    
    ## The for loop pauses or sleeps for 10 seconds to mimic human interaction with the website
    time.sleep(10)
```

## Loop to parse different HTML files saved into a BeautifulSoup object and then printing out the links/URLs of the various sponsored items on display

```python
for i in range (1,11):
    ## Setting the file name
    file_name = "ebay_ps4_"+"%02d"%i+".htm"

    with open(file_name, "rb") as html_file:
        file_text = html_file.read()

        ## Parsing the files into a BeautifulSoup object
        soup = BeautifulSoup(file_text, 'html.parser')

        ## Finding tag that contains all the results
        items = soup.select("#srp-river-results > ul > li.s-item")            

        ## Iterating through the above tag to find all <a> tags in it
        for item in items:
            a = item.find("a")                
            if not a:
                continue                


            ## If it is <a> tag, then checking whether the span tag in it contains "SPONSORED" in the given way because of the html code
            text = item.find("span").get_text()                
            if re.search('.*S.*P.*O.*N.*S.*O.*R.*E.*D.*',text):
                
                ## Shortening the URL because the part after the ? does not matter
                url_simplify = re.sub("(.*)\\?.*", r"\1", a['href'])

                ## Saving the Sponsored URLs in a file called "sponsored.txt"

                save_file = open("sponsored.txt", "a")
                save_file.write(url_simplify+"\n")
                save_file.close()

                
            ## If the span tag does not contain "SPONSORED" then it goes to this loop
            else:
                ## Shortening the URL because the part after the ? does not matter
                url_simplify = re.sub("(.*)\\?.*", r"\1", a['href'])

                ## Saving the Non-Sponsored URLs in a file called "non_sponsored.txt"

                save_file = open("non-sponsored.txt", "a")
                save_file.write(url_simplify+"\n")
                save_file.close()
                

```

```python
list_of_files_sponsored = os.listdir("sponsored")
```

```python
from tqdm import tqdm
```

```python
sponsored_item_data = []

for file_name in tqdm(list_of_files_sponsored):
    
    try:
    
        html_file = open("./sponsored/"+file_name, "rb")
        file_text = html_file.read()
        html_file.close

        item_number = int((file_name.split("."))[0])

        soup = BeautifulSoup(file_text, 'html.parser')

        ## Seller Name - Selector used is span and class
        seller_name = (soup.find('span', attrs = {'class':'mbg-nw'})).text

        ## Seller Score - Selector used is span and class
        seller_score = int(((soup.find('span', {'class':'mbg-l'})).find('a')).text)

        ## Item price - Selector used is span and itemprop
        item_price = int((float((soup.find('span', {'itemprop':'price'}))['content']))*100)

        ## Number of items sold - Selector used is div and ID
        item_sold_div = soup.find('div', {'id':'mainContent'})
        if (re.search("(\d+) sold",(item_sold_div.text))):
            n_sold = re.search("(\d+) sold",(item_sold_div.text))
            items_sold = int(float(n_sold.groups()[0]))
        else:
            items_sold = None

        ## Best offer available - Selector used is form and name
        best_offer_form = soup.find('form', {'name':'viactiondetails'})
        if re.search("Best Offer",(best_offer_form.text)):
            best_offer = True
        else:
            best_offer = False

        ## Title - Selector used is h1 and ID
        title = str((soup.find('h1', {'id':'itemTitle'})).find(text=True, recursive=False))

        ## Returns allowed - Selector used is div and ID
        returns_allowed_tag = soup.find('div', {'id':'why2buy'})
        if re.search("Returns accepted",(returns_allowed_tag.text)):
            returns_accepted = True
        else:
            returns_accepted = False

        ## Shipping price - Selector used is span and id
        shipping_price_parent = soup.find('span', {'id':'shSummary'})
        if re.search('free', shipping_price_parent.text, flags=re.IGNORECASE):
            shipping_price = int(0)
        elif re.search(r'\$[\d.]+', shipping_price_parent.text):
            shiping_price_tag = re.search("\$([\d.]+)",(shipping_price_parent.text))
            shipping_price = int((float(shiping_price_tag.groups()[0]))*100)
        else:
            shipping_price = None

        ## Condition - Selector used is div and ID
        condition = (soup.find('div', {'id':'vi-itm-cond'})).text
        
        ## Sponsored or Non-Sponsored
        spons = True

        data = [item_number, 
                spons, 
                seller_name, 
                seller_score, 
                item_price, 
                items_sold, 
                best_offer, 
                title, 
                returns_accepted, 
                shipping_price, 
                condition]

        sponsored_item_data.append(data)
    
    except:
        continue
```
